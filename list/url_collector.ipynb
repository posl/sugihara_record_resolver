```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import re

language = 'Java'
stars = 1000
pushed = '2020-03-17'

loop_limit = 10

def typeA_URL(page, pushed, language, stars_limit):
  return f"https://github.com/search?o=desc&p={page}&q=pushed%3A%3E%3D{pushed}+and+language%3A{language}+and+stars%3A%3E%3D{stars_limit}&s=stars&type=Repositories"

def typeB_URL(page, pushed, language, stars_num):
  return f"https://github.com/search?o=desc&p={page}&q=pushed%3A%3E%3D{pushed}+and+language%3A{language}+and+stars%3A{stars_num}&type=Repositories"

def typeC_URL(pushed, language, stars_num):
  return f"https://github.com/search?o=desc&q=pushed%3A%3E%3D{pushed}+and+language%3A{language}+and+stars%3A{stars_num}&type=Repositories"

def typeD_URL(page, pushed, language, lower_bound, upper_bound):
  return f"https://github.com/search?o=desc&p={page}&q=pushed%3A%3E%3D{pushed}+and+language%3A{language}+and+stars%3A{lower_bound}..{upper_bound}&type=Repositories"

def to_repository_URL(results, counter):
  links = []
  for req in results:
    repo_name = ""
    for rep in req.contents:
      if(hasattr(rep, 'text')):
        repo_name = repo_name + str(rep.text)
      else:
        repo_name = repo_name + rep
    link = f"https://github.com/{repo_name}"
    counter.count += 1
    print(f"{counter.count} : {link}")
    links.append(link)
  return links

def extract_repository(timeout, url):
  r = requests.get(url)
  soup = BeautifulSoup(r.text,'html.parser')
  time.sleep(timeout)
  results = soup.find_all("a", class_="v-align-middle")
  loop = 0
  while(len(results) == 0 and loop < loop_limit):
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    time.sleep(timeout * (loop/2 + 1))
    results = soup.find_all("a", class_="v-align-middle")
    loop += 1
  return results

class Counter:
  count = 0

counter = Counter()

def collect_repositories_A(language, pushed, sort, timeout):
  links = []
  for i in range(100):
    results = extract_repository(timeout, typeA_URL(i+1, pushed, language, 1000))
    if(len(results) == 0):
      print("no response")
      print(f"check {typeA_URL(1, pushed, language, 1000)}")
      break
    links += to_repository_URL(results, counter)

  return links

def collect_repositories_B(language, pushed, sort, timeout, lower_bound, upper_bound):
  links = []
  for i in range(100):
    results = extract_repository(timeout, typeD_URL(i+1, pushed, language, lower_bound, upper_bound))
    if(len(results) == 0):
      print("no response")
      print(f"check {typeD_URL(i+1, pushed, language, lower_bound, upper_bound)}")
      break
    links += to_repository_URL(results, counter)
  return links

links_A = collect_repositories_A('java', '2020-03-17', 'desc', 4)
links_B = collect_repositories_B('java', '2020-03-17', 'desc', 4, 400, 999)
links_C = collect_repositories_B('java', '2020-03-17', 'desc', 4, 250, 399)

for link in links_A:
  print(link)
for link in links_B:
  print(link)
for link in links_C:
  print(link)
```

```python
def extract_repository_no_retry(timeout, url):
  r = requests.get(url)
  soup = BeautifulSoup(r.text,'html.parser')
  time.sleep(timeout)
  return soup.find_all("a", class_="v-align-middle")

def collect_repositories_B(language, pushed, sort, num, timeout):
  links = []
  for i in range(1000):
    results = extract_repository_no_retry(timeout, typeC_URL(pushed, language, 999-i))
    if(len(results) == 10):
      for j in range(100):
        links += to_repository_URL(extract_repository(typeB_URL(j+1, pushed, language, 999-i)), counter)
    else:
      links += to_repository_URL(results, counter)
    if(counter.count >= num):
      break
  return links

links_B = collect_repositories_B('java', '2020-03-17', 'desc', 2300-len(links_A), 10)
for link in links_B:
  print(link)
```
